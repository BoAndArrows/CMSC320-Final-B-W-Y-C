```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from Sine_With_Poly_Fit import SineWithPolyModel
```

```python
samples = 1000
X = np.linspace(0, 6*np.pi, samples).reshape(-1, 1)
Y = 10*np.sin(X) + 0.0003*X**3 + 0.0002*X**2 + 0.0005*X + 1
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=False)

model = SineWithPolyModel(degree=3, iterations=5000, lr = 0.0001).fit(X_train.reshape(-1, 1), Y_train)

def plot_regression_line(model, X, Y):
    # Predict values using the fitted model
    X = X #.reshape(-1, 1)
    Y_pred = model.predict(X)

    # Plot the actual data and the fitted regression line
    plt.scatter(X, Y, color='blue', label='Actual Data')
    plt.plot(X, Y_pred, color='red', label='Fitted Line')
    plt.title('Traffic Volume Prediction Using Regression')
    plt.xlabel('Time of Day (hours)')
    plt.ylabel('Traffic Volume (vehicles/hour)')
    plt.legend()
    plt.show()

plot_regression_line(model, X_train, Y_train)
```
